## Filling Null values with "Unknown" and ' ' .

```python
df.fillna({'Movie_Director': 'Unknown',
          'Movie_Tagline': 'No Tagline Available',
          'Movie_Cast':'Unknown',
          'Movie_Keywords':'',
          'Movie_Homepage':''}, inplace=True)
```

## Again Check Null Values

```python
df.isna().sum()[df.isna().sum() > 0]
```

## OUTLIERS IDENTIFICATION

```python
top_movies = df[['Movie_Title', 'Movie_Revenue']].sort_values(
    by='Movie_Revenue', ascending=False
).head(5)
top_movies
```

## Visualization 

```python
plt.pie(
    top_movies['Movie_Revenue'],
    labels=top_movies['Movie_Title'],
    startangle=100
)

plt.title('Top 5 Movies by Revenue')
plt.show()
```
